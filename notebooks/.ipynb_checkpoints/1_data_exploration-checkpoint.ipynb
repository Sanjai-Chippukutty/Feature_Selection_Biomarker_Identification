#  Notebook 1 â€” Data Exploration (EDA)

##  Abstract
This notebook performs **Exploratory Data Analysis (EDA)** on the **Wisconsin Breast Cancer Diagnostic Dataset** to understand its structure, detect potential data quality issues, and identify important statistical patterns.

The objectives of this EDA are to:
- Understand the **size, structure, and data types** of the dataset.
- Inspect **class distribution** for balanced or imbalanced labels.
- Explore **feature distributions** for patterns and anomalies.
- Examine **correlations** between features to detect redundancy.


##  Importing Required Libraries

Before starting any data analysis or visualization, we first import the necessary Python libraries.


```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid", palette="pastel", font_scale=1.2)
```

##  Loading the Dataset

We will now load our dataset into a Pandas DataFrame for analysis.


```python
# Load dataset
data_path = "C:/Users/sanja/2.Feature_Selection_Biomarker_Identification/Feature_Selection_Biomarker_Identification/data/raw/breast-cancer.csv"
df = pd.read_csv(data_path)

# Display first 5 rows
df.head()
```

##  Dataset Overview & Summary Statistics

After loading the dataset, the next step is to **understand its structure** and **basic statistical properties**.


```python
# Dataset overview
df.info()

# Summary statistics
df.describe().T
```

##  Checking for Missing Values

Before proceeding with data preprocessing and modeling, it is important to check if the dataset contains **missing values**, as they can affect analysis and model performance.


```python
# Check missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

##  Target Class Distribution

Understanding the **distribution of the target variable** is an important step before training any classification model.  
Here, our target variable is **`diagnosis`**, which indicates whether a tumor is **malignant (M)** or **benign (B)**.


```python
# Class distribution plot
sns.countplot(data=df, x='diagnosis', palette="Set2")
plt.title("Target Class Distribution")
plt.xlabel("Diagnosis (M = Malignant, B = Benign)")
plt.ylabel("Count")
plt.show()

# Percentage distribution
class_distribution = df['diagnosis'].value_counts(normalize=True) * 100
class_distribution
```

##  Feature Distributions â€” Histograms

Before applying machine learning algorithms, it is useful to **understand the distribution of each feature**.  
Histograms allow us to see **how values are spread** across the dataset for each numerical feature.

```python
# Histograms for all numerical features
df.hist(figsize=(18, 14), bins=30, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=20)
plt.show()
```

##  Feature Correlation Analysis â€” Heatmap

Correlation analysis helps identify **relationships between numerical features** in the dataset.  
Highly correlated features can provide redundant information, which may be addressed during **feature selection**.


```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Feature Correlation Heatmap", fontsize=18)
plt.show()
```
